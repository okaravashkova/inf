```python
'''
Контрольные вопросы:

1) Как расшифровывается СКВ (VCS) и зачем они нужны в больших проектах?

Системы контроля версий (Versions control systems) - ПО для хранения часто обновляющейся информации, например, для хранения исходного кода какого-то ПО.
В проектах, в которых задействовано много человек, так удобно отслеживать кто какие изменения внёс.

2) В чем отличия централизованных СКВ от децентрализованных?

В децентрализованных СКВ у каждого клиента есть копия всех данных, в централизованных - весь объём данных хранится на главном сервере.


3) Что такое git и чем он отличается от GitHub?

Гит - система контроля версий. Гитхаб - онлайн-сервис для проектов, использующих git.

4) Какие основные команды используются при работе с git? Что они делают?

4.1 `git commit -m <message>`: Создает коммит с сохраненными изменениями и сообщением, описывающим изменения.

4.2 `git pull`: Получает изменения с удаленного репозитория и сливает их с текущей веткой.

4.3 `git push`: Отправляет коммиты из локального репозитория в удаленный репозиторий.

4.5 `git merge <branch>`: Объединяет указанную ветку с текущей веткой.


5) Зачем добавлять файлы в gitignor?

Чтобы они не попадали под git-команды и не подтягивались в репозиторий.

6) В каких случаях имеет смысл создавать различные ветки проекта?

Если нужно создать какую-то новую функцию в приложении, например (или пофиксить какой-то баг или ), 
то можно создать новую ветку под неё, а после окончания работы добавить то, что мы сделали в основную ветку.
Это дает возможность легко откатывать код, если вдруг передумаем его сливать в основную ветку, 
либо делать несколько различных изменений в разных ветках.

7) В чем отличия merge и rebase? В чем преимущества и недостатки последнего из них?

Merge объединяет изменения двух веток, создавая новый коммит с объединенными изменениями. 
Merge сохраняет историю коммитов обеих веток и позволяет объединять изменения из разных веток.

Rebase же перемещает коммиты одной ветки поверх другой ветки. 
Результатом является новая история коммитов, в которой коммиты из обеих веток идут последовательно.

Плюсы: 
- более лаконичная и структурированная история изменений в проекте
Минусы:
- меняется история коммитов, что  может привести к проблемам, если другие разработчики уже использовали старую историю коммитов.
- уже не отследить историю до слияния
- могут возникнуть ошибки при откате

'''
```
