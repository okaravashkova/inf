```python
 '''5. Как порядок декорирования влияет на результат работы функции? 

Сначала выполняется самый ближайший к функции декоратор:'''

def makebold(fn):
    def wrapped():
        return "<b>" + fn() + "</b>"
    return wrapped
 
def makeitalic(fn):
    def wrapped():
        return "<i>" + fn() + "</i>"
    return wrapped
 
@makebold
@makeitalic
def hello():
    return "hello copypast..."
 
print hello() ## выведет <b><i>hello copypast...</i></b>
```
