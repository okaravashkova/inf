```python
class Complex:
    def __init__(self, real = 0, im = 0):
        self.real = real
        self.im = im
    def set_numb(self, real, im):
        self.real = real
        self.im = im
    def get_real(self):
        return self.real
    def get_im(self):
        return self.im
    def __str__(self):
        return f'Result = {self.real} + {self.im}*i'

def add(A: Complex, B: Complex):
    C = Complex()
    C.set_numb((A.get_real()+B.get_real()), (A.get_im()+B.get_im()))
    print(C.__str__())

def substr(A: Complex, B: Complex):
    C = Complex()
    C.set_numb((A.get_real()-B.get_real()), (A.get_im()-B.get_im()))
    print(C.__str__())

def div(A: Complex, B: Complex):
    C = Complex()
    x1 = A.get_real()
    x2 = B.get_real()
    y1 = A.get_im()
    y2 = B.get_im()
    C.set_numb((x1*x2 + y1*y2)/(x2**2 + y2**2), (x2*y1 - x1*y2)/(x2**2 + y2**2))
    print(C.__str__())

def multi(A: Complex, B: Complex):
    C = Complex()
    x1 = A.get_real()
    x2 = B.get_real()
    y1 = A.get_im()
    y2 = B.get_im()
    C.set_numb((x1*x2 - y1*y2), (x1*y2 + y1*x2))
    print(C.__str__())


A = Complex()
B = Complex()
A.set_numb(int(input('Введите действительную часть первого числа')), int(input('Введите мнимую часть первого числа')))
s = str(input('Выберете операцию: +, -, *, :'))
B.set_numb(int(input('Введите действительную часть второго числа')), int(input('Введите мнимую часть второго числа')))

if s == '+':
    add(A, B)
elif s == '-':
    substr(A, B)
elif s == '*':
    multi(A, B)
elif s == ':':
    div(A, B)
else:
    print('Ошибка ввода')
```
